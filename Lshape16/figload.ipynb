```python
fig, axs = plt.subplots(10, 11, figsize=(15, 15))
plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
plt.subplots_adjust(wspace=0,hspace=0)
for GEn in range(10):
    for HEn in range(11):
        el = GEn*dge
        hev = HEn*dhev
        img = imread(fname=result_dir+f'/el_{el}_ge_{ge}_hev_{hev}_dc_160.0.png', format='ping')
        axs[GEn][HEn].imshow(img)
        axs[GEn][HEn].tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
        axs[GEn][HEn].tick_params(bottom=False,left=False,right=False,top=False)
        axs[GEn][HEn].spines['right'].set_visible(False)
        axs[GEn][HEn].spines['top'].set_visible(False)
        axs[GEn][HEn].spines['bottom'].set_visible(False)
        axs[GEn][HEn].spines['left'].set_visible(False)
plt.show()
```

```python
obj  = np.load(result_dir + '/obj_list.npy',allow_pickle=True)
frac = np.load(result_dir + '/frac_list.npy',allow_pickle=True)
heav = np.load(result_dir + '/heav_list.npy',allow_pickle=True)
```

```python
ax = plt.axes()
cm = plt.get_cmap('jet')
eloss = np.zeros(110)
for i in range(len(obj)):
    eloss[i]=(np.array(obj[i])[-1])
plt.imshow(eloss.reshape((10,11)),cmap='jet',vmax =14.9)
ax=plt.gca()
ax.set_xticks(np.arange(0, 11, 1))
ax.set_yticks(np.arange(0, 10, 1))
# ax.set_xticklabels(np.arange(0, 100, 10))
# ax.set_yticklabels(np.arange(0, 440, 40))
ax.set_xticks(np.arange(-.5, 11, 1), minor=True)
ax.set_yticks(np.arange(-.5, 10, 1), minor=True)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
ax.grid(which='minor', color='w', linestyle='-', linewidth=1.5)
plt.colorbar()
plt.tick_params(bottom=False,left=False,right=False,top=False)
plt.show()
print(np.amin(eloss),np.amax(eloss))
```

```python
ax = plt.axes()
cm = plt.get_cmap('jet')
err = np.zeros(110)
for i in range(len(heav)):
    err[i]=(np.array(heav[i])[-1])
plt.imshow(err.reshape((10,11)),cmap='jet',vmax=14.9)
ax=plt.gca()
ax.set_xticks(np.arange(0, 11, 1))
ax.set_yticks(np.arange(0, 10, 1))
# ax.set_xticklabels(np.arange(0, 110, 10))
# ax.set_yticklabels(np.arange(0, 400, 40))
ax.set_xticks(np.arange(-.5, 11, 1), minor=True)
ax.set_yticks(np.arange(-.5, 10, 1), minor=True)
ax.grid(which='minor', color='w', linestyle='-', linewidth=1.5)
plt.colorbar()
plt.tick_params(bottom=False,left=False,right=False,top=False)
ax.tick_params(labelbottom=False,labelleft=False,labelright=False,labeltop=False)
plt.show()
print(np.amin(err),np.amax(err))
```

```python
ax = plt.axes()
cm = plt.get_cmap('jet')
erfrac = np.zeros(110)
for i in range(len(frac)):
    erfrac[i]=np.abs((np.array(frac[i])[-1]-0.5))
plt.imshow(erfrac.reshape((10,11)),cmap='jet')
ax=plt.gca()
ax.set_xticks(np.arange(0, 11, 1))
ax.set_yticks(np.arange(0, 10, 1))
ax.set_xticklabels(np.arange(0, 110, 10))
ax.set_yticklabels(np.arange(0, 400, 40))
ax.set_xticks(np.arange(-.5, 11, 1), minor=True)
ax.set_yticks(np.arange(-.5, 10, 1), minor=True)
ax.grid(which='minor', color='w', linestyle='-', linewidth=1.5)
plt.colorbar()
plt.tick_params(bottom=False,left=False,right=False,top=False)
plt.show()
print(np.amin(erfrac),np.amax(erfrac))
```
